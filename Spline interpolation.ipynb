```python
import numpy as np
from scipy.interpolate import BSpline
import matplotlib.pyplot as plt

# Define the knots and coefficients
knots = np.array([0, 1, 2, 3, 4, 5])
coefficients = np.array([0, 1, 0, -1, 0, 1])
degree = 3

# Create the spline
spline = BSpline(knots, coefficients, degree)

# Generate some data points
x = np.linspace(0, 5, 100)
y = spline(x)

# Plot the spline
plt.plot(x, y, label='BSpline')
plt.scatter(knots, coefficients, color='red', label='Knots')
plt.legend()
plt.show()
```
